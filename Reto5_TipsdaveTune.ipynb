## ¿Razón?

**Definición**: La razón entre un número **a** y un número **b** es el cociente los mismos.

${razón} = \frac{a}{b}$

## Entonces...

${razón} = \frac{Número Total de casos COVID-19}{Número Total de camas de hospital disponibles}$

## ¿Como obtengo el número total de camas de hospital disponibles?

-> Hallar el número total de camas de hospital por **1 persona**

$ {Número de camas de hospital por persona} = \frac{Número Total de camas de hospital disponibles}{1000} $

-> Hallar el número total de camas para toda la población de determinado país

$ {Número total camas} = {Número de camas de hospital por persona}*{población}$

## ¿Como Operar columnas de un Dataframe?

```python
datosLacteos
```

**Ejemplo:** Vamos a multiplicar la cantidad en ventas de un producto por el valor del mismo

```python
# Creamos una nueva columna para almacenar los valores que vamos a operar
datosLacteos['total_ventas'] = datosLacteos['valor_unidad_producto'] * datosLacteos['cantidad_vendida']
datosLacteos
```

## ¿Como agrupamos y obtenemos datos de dicha agrupación?

**Función de pandas =>  groupby**

### Estructura:

nombredataframe.groupby(**[por cual columna queremos agrupar]**)[**que valor queremos rescatar**].**operacion para todos los valores que encuentre en ese grupo**

### ¿Que operaciones puedo usar?



```python
# Agrupamos por el nombre del vendedor y vamos a obtener la suma del total de ventas que hizo cada  vendedor
group_V_TV = datosLacteos.groupby(['vendedor'])['total_ventas'].sum()
group_V_TV
```

## ¿Como agrupamos por 2 columnas?

Agregamos el parametro **by** dentro de la función **groupby** y dentro de una lista le ponemos los nombres de las columnas que queremos agrupar.

```python
# Agrupamos por el nombre del vendedor y nombre del producto y vamos a obtener la suma del total de ventas 
# que hizo cada  vendedor
group2_V_TV = datosLacteos.groupby(by = ['vendedor', 'producto'])['total_ventas'].sum()
group2_V_TV
```

### ¿Como separo las columnas agrupadas?

Hacemos uso de la función **reset_index()**

```python
# Agrupamos por el nombre del vendedor y nombre del producto y vamos a obtener la suma del total de ventas 
# que hizo cada  vendedor
group2_V_TV = datosLacteos.groupby(by = ['vendedor', 'producto'])['total_ventas'].sum().reset_index()
group2_V_TV
```

## ¿Como organizamos el DataFrame agrupado, para que sea más legible?

Hacemos uso de la función **pivot**

### Estructura:

nombredataframe.**pivot**(index = 'datos que se pondran en las filas', columns = 'datos que se pondran en las columnas',
values = 'valores del dataframe')


```python
pivotDt = group2_V_TV.pivot(index='vendedor', columns='producto', values='total_ventas')
pivotDt
```
